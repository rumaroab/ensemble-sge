```python
import json
import matplotlib.pyplot as plt
import numpy as np

def readFile(name = "res.json"):
    return json.load(open(name))
```

```python
title = "Pagie Polinomial full"
grname = "pagie_poli"

filename = "symbolic_pagie_poli_0_1"
filename1 = "symbolic_pagie_poli_0_0"
filenamefull = "symbolic_pagie_poli_0_0"

data = []
for obj in readFile("results/v2/"+filename+".json"):
    if obj["ensemble"] > 4:
        obj["ensemble"] = np.nan
    data.append(obj)
data1 = []
for obj in readFile("results/v2/"+filename1+".json"):
    data1.append(obj)
datafull = readFile("results/v2/"+filenamefull+".json")

plt.title(title)
#plt.plot([data[i]["ensemble"]for i in range(len(data))], color="red", label='Ensemble SGE Mean')
plt.plot([data1[i]["ensemble"]for i in range(len(data))], color="orange", label='Ensemble SGE Median')
plt.plot([datafull[i]["best"]for i in range(len(datafull))], color="green", label='SGE')
plt.legend()

#plt.ylabel([data[i]["generation"] for i in range(len(data)) ])    
#plt.savefig('results/plots/'+grname+'.png')

plt.show()
```
