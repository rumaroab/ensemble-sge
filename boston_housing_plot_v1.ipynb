```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python

def readFile(name = "res.json"):
    return json.load(open(name))

```

```python
title = "Boston full "
grname = "boston_1000"

filename = "boston_housing_0_1_v1"
filename1 = "boston_housing_0_0_v1"
filenamefull = "boston_housing_0_0_v1"
data = []
for obj in readFile("results/v2/"+filename+".json"):
    if obj["ensemble"] > 4:
        obj["ensemble"] = np.nan
    data.append(obj)
data1 = []
for obj in readFile("results/v2/"+filename1+".json"):
    data1.append(obj)
datafull = readFile("results/v2/"+filenamefull+".json")

plt.title(title)
plt.plot([data[i]["ensemble"]for i in range(len(data))], color="red", label='Ensemble Mean')
plt.plot([data1[i]["ensemble"]for i in range(len(data))], color="orange", label='Ensemble Median')
plt.plot([data[i]["best"]for i in range(len(data))], color="green", label='SGE')
plt.plot([datafull[i]["best"]for i in range(len(datafull))], color="green", label='SGE')
plt.legend()
#plt.ylabel([data[i]["generation"] for i in range(len(data)) ])    
#plt.savefig('results/plots/'+grname+'.png')

plt.show()
```
